```
import numpy as np
import pylab as plt
from scipy.optimize import minimize
```

```
x = np.arange(10)
```

```
def yfunc(a, b):
  global x
  return a*x+b
```

```
y = yfunc(2.3, 4.5) + np.random.normal(0, 1, 10)
yerr = np.random.normal(0, 1, 10)
```

```
plt.errorbar(x, y, yerr=yerr, fmt='*')
```

```
def func(p):
  a = p[0]
  b = p[1]
  y1 = yfunc(a, b)
  ls = np.sum((y - y1)**2/yerr**2)
  return ls
```

```
print (func((3, 5)))
```

```
x0 = (1,5)
m = minimize(func, x0)
```

```
print (m)
```

```
print (m)
```

```
yfit = m['x']
plt.plot(x, y, '*')
plt.plot(x, yfunc(*yfit))
```

```
yfit = m['x']
plt.errorbar(x, y, yerr=yerr, fmt='*')
plt.plot(x, yfunc(*yfit))
```
